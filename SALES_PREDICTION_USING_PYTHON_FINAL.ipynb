```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style
sns.set_style('whitegrid')

# Define custom colors
colors = ['red', 'green', 'blue']

# Create the figure and axes
fig, axes = plt.subplots(ncols=3, figsize=(12, 4))

# Scatter plot for TV
sns.scatterplot(data=test_data, x='TV', y='Sales', ax=axes[0], color=colors[0])
axes[0].set_title('TV')

# Scatter plot for Radio
sns.scatterplot(data=test_data, x='Radio', y='Sales', ax=axes[1], color=colors[1])
axes[1].set_title('Radio')

# Scatter plot for Newspaper
sns.scatterplot(data=test_data, x='Newspaper', y='Sales', ax=axes[2], color=colors[2])
axes[2].set_title('Newspaper')

# Show the plot
plt.show()
```

```python
X = test_data.drop('Sales', axis=1)
y = test_data['Sales']
```

```python
X1 = test_data[['TV', 'Radio']]
X2 = test_data[['TV', 'Radio', 'Newspaper']]
```

```python
from sklearn.model_selection import train_test_split

for X, name in zip([X1, X2], ['TV and Radio only', 'All features']):

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn import metrics

y_predictions = model.predict(X_test)
r2 = metrics.r2_score(y_test, y_predictions)
print(f'{name} - R-Squared:', r2)
```
