```python
train_Data = train_Data.rename(columns={train_Data.columns[0]:'State',train_Data.columns[3]:'EUR',train_Data.columns[4]:'EE', train_Data.columns[5]:'ELPR', train_Data.columns[6]:'Region'})
train_Data.head()
```

```python
reg_tics = train_Data.groupby(['Region'])[['EE','EUR','ELPR']].mean().reset_index()
reg_tics = round(reg_tics,2)
reg_tics
```

```python
heat_maps = train_Data[['EE', 'EUR', 'ELPR',  'latitude','longitude']]
heat_maps = heat_maps.corr()

plt.figure(figsize=(10, 6))
sns.set_context('notebook', font_scale=1)
sns.heatmap(heat_maps, annot=True, cmap='magma')
```

```python
train_Data.columns = ["State", "Date", "Frequency", "EUR", "EE", "ELPR", "Region", "longitude", "latitude"]

plt.figure(figsize=(10, 8), dpi=1300)
plt.title("Unemployment Rate")
sns.histplot(x="EUR", hue="Region", data=train_Data, palette=['blue', 'green', 'red', 'purple', 'yellow'])
plt.show()
```

```python
import plotly.express as px
```

```python
import plotly.express as px

region = train_Data.groupby(["Region"])[['EUR', 'EE', 'ELPR']].mean()
region = pd.DataFrame(region).reset_index()

# Define custom color order
color_order = ['green', 'blue', 'red', 'yellow', 'purple']  # Adjust the colors as desired

fig = px.bar(region, x="Region", y="EUR", color="Region", title="Region By Average Unemployment Rate",
             color_discrete_sequence=color_order)
fig.update_layout(xaxis={'categoryorder': 'total descending'})
fig.show()
```

```python
unemployment = train_Data[["State", "Region", "EUR"]]

fig = px.sunburst(unemployment, path=['Region', 'State'], values='EUR',
                  title='Unemployment Rate by State and Region  by circular chart', height=650,
                  color_discrete_sequence=['blue', 'green', 'red', 'purple', 'yellow'])
fig.show()
```
